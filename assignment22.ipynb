Questions
Questions you must answer about your conceptual model:
1. What are the ranges, data types and format of all of the attributes in your entities?
VARCHAR(5, 10), INT, TIME, FLOAT 

2. When should you use an entity versus attribute? (Example: address of a person could be modeled as either)
Since the main difference between Entity and Attribute is that an entity is a real-world object that represents data in RDBMS while an attribute is a property that describes an entity. When we would like to decribe an address, we take it as an entity and then use the city, street...etc as its attribute. When we take it as a information as a person, it is going to be an attribute.

3. When should you use an entity or relationship, and placement of attributes? (Example: a manager could be modeled as either)
A manager can be an entity or an attribute accroding to different situations or tables. When we decribe the attributes of a manager like their clints or branches, we take it as an entity. If we make a employee table for the compant the manager will be an attribute.

4. How did you choose your keys? Which are unique?
We choose the player_id, Season, team_id and coach_id as our primary keys. And the team_id in Coach and Player tables are the foreign key. 

5. Did you model hierarchies using the “ISA” design element? Why or why not?
Yes, the relationship between player table and performance table we chose“ISA”, because their relationship is inheritances. The class “performance” which inherit is known as sub-class. That means the performance of player is also a type of player’s information, like an apple is a fruit, so we extend player’s information to get the performance. Other relationship we chose “HASA”, because from table to table they have composite relations, that means the use of instance variables that are references to other objects. Such as the team has a coach, player has a team,etc.

6. Were there design alternatives? What are their tradeoffs: entity vs. attribute, entity vs. relationship, binary vs. ternary relationships?
Yes, there are alternatives. We can put the performacne into the players table. But considered that the performance will change by season, we make it an individual table. 

7. Where are you going find real-world data populate your model?
We can add battle information which is the data between two teams in a battle.And also add more basic information about players like salary, nickname,etc... And also the player performance from every season.

Questions about physical model: 

1. Are all the tables in 1NF?
Yes, our original data sets are in 1NF. 
Each table has a primary key: minimal set of attributes which can uniquely identify a record. 
The values in each column of a table are atomic (No multi-value attributes allowed).
There are no repeating groups: two columns do not store similar information in the same table.

2. Are all the tables in 2NF?
Yes, our original data sets are in 2NF.
All requirements for 1st NF has been met.
No partial dependencies.
No calculated data

3. Are all the tables in 3NF?
Yes, our original data sets are in 3NF.
All requirements for 2nd NF has been met.
All fields directly depend on the primary key; that is no transitive dependencies.
